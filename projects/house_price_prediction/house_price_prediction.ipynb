```python
import pandas as pd
```

```python
data = pd.read_csv('housing.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
mainroad_en = le.fit_transform(data.loc[:,'mainroad'])
```

```python
# data['mainroad_en'] = mainroad_en
```

```python
categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 
                   'airconditioning', 'parking', 'prefarea', 'furnishingstatus']
```

```python
for cols in categorical_cols:
    data[cols] = le.fit_transform(data[cols])
```

```python
data
```

```python
X = data.iloc[:,1:]
```

```python
X
```

```python
y = data.loc[:,['price']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test, y_train, y_test = tts(X,y, train_size = 0.8, random_state = 44)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
```

```python
mae
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
rs = r2_score(y_test, y_pred)
```

```python
rs
```

```python
import joblib
```

```python
joblib.dump(model, 'housing.pkl')
```
